### Web Foundation

Surfing the internet is basically interacting with an always-on web server from a client terminal. The browser sends a request message to the internet service provider (ISP), which liaises with the domain name server (DNS) to look up the [exact IP address](https://www.nslookup.io) of the requested website. Thereafter, you can make direct requests to that specific IP address for the content hosted there. Modern servers use dynamic IP addresses (changing based on usage) and a content distribution network (geographically distributed addresses to lower latency). Messages are sent across the world via optical fibres in [large underwater cables](https://www.submarinecablemap.com) at the speed of light.

#### HTML, CSS and JavaScript

To access a webpage, a browser looks up and receives data provided by a server. The data primarily consists of three code files: HTML, CSS and JavaScript.<br><br>
The <b>.HTML</b> establishes the qualitative structure of a website, such as images, buttons, or textboxes. Early websites, like the [first one to be created](https://info.cern.ch/hypertext/WWW/TheProject.html), contained HTML without any CSS or JavaScript. Conversely, however, it is not possible to have only the other two without a HTML file<br>
The <b>.CSS</b> stylises the aesthetics of the website, specifying the colour of the button, or the font of the text.<br>
The <b>.JavaScript</b> contains the functionality or behaviour of the website.

Then, the browser renders each of these files as a website

#### HyperText Markup Language

HyperText are pieces of text that can link to another documents in the website. [HTML tags](https://www.w3schools.com/TAGS/default.asp) ([Link 2](https://developer.mozilla.org/en-US/docs/Web/HTML/Guides/Cheatsheet)) are used to indicate the desired formatting to apply to the text. The tag + text is referred to as an element. Examples include the header element, in which the levels are inspired by book bindings, with top level headings and subsections (H1 through to H6), paragraph tags (for which [placeholder text](https://lipsum.com) are used to test formatting), unordered and ordered lists (with list items). There are also void elements, where content is forbidden within the tags, such as the horizontal separator "hr" and line break "br" tags (both enclosed in <>). Line breaks are used for natural new lines such as for poems, addresses etc.

#### HTML Boilerplate

HTML has several elements that are included with each document:
1. "!DOCTYPE html" declaration, which indicates to browsers the html version (version 5)
2. "html" start and end tags, the root of the document enclosing the entire document
3. html lang="en", the language of the content for screen readers
4. head, which includes important information about the website for rendering, but does not contain anything structurally. head contains the meta charset="UTF-8", the character set for encoding of the webpage, and title, the title displayed in the tab bar. meta name"viewport" content="width=device-width, initial-scale=1.0", which denotes the size of the window when the website is first opened.
5. body, in which the proper content of the website is nested (text, titeles, images, links, etc.)

#### Cascading Style Sheets

CSS is used to [style websites](https://appbrewery.github.io/just-add-css/), when before only HTML was available, in ways that were "trivial to other documents like in TeX or Microsoft Word". W3 first introduced font, colour, face, size and center tags were introduced in HTML 3.2. However, they became deprecated as these were considered non-essential for content and cluttering. In 1996, the W3 consortium (Håkom Wium Lie) proposed the Cascasing Style Sheets, which "attached style" to HTML documents in a modular (and thus delegable) manner. CSS, Sass, and Less are all members of style sheet languages that apply styles.

There are 3 ways of adding CSS to a HTML website:

1. Inline: goes into the opening tag of an HTML element as "style=", a global attribute, which takes a CSS code in a "property: value" format. Surgical and not recommended
2. Internal: CSS code is added inside of style tags. Inside the tags, the internal CSS takes a HTML selector "html", before the specific code goes into curly braces. Only applies to the specific webpage the internal method is included in.
3. External: The sme HTML selector and CSS code (property: value, separated by ; and enclosed by curly braces), lives in a separate styles.css document, which is referenced by the main HTML document in the head, within a self closing "link" tag. It takes rel="stylesheet", denoting the relationship to the document, and "href="./styles.css", the location of the styles.css file. Applies globally to the website and most commonly used in web development.

The external method applies styles in a "cascading" manner, from the most general to specific styling requirements, the latter achieved by using CSS Selectors to target the specific section. CSS Selectors include:

1. Universal Selector
2. Generic Element Selectors (h1, p, ol as above)
3. Class Selectors: Classes are attributes that are used to group HTML elements, in order to be referenced to apply the same styles on (like tagging elements and applying style to elements with that tag)
4. Id Selectors: Applied on unique element throughout the webpage assigned with that Id (equivalent to having single inline, but without mixing content structure with presentation)
5. Attribute Selectors: Select a subset of elements that have a specific attribute applied to it


````python
/* universal selector */
* {
    color: red;
}

/* element selector on all header 2 elements */
h2 {
    color: blue;
}

/* class selector selecting those given class="red-heading" */
.red-heading { 
    color: red;
}

/* id selector for element with id="main" */
#main {
    color: green;
}

/* attribute selector: all paragraph elements with the draggable attribute */
p[draggable]{
    color: red;
}

/* further, it can select elements in a specific state of specified attribute */
p[draggable="false"]{
    color: red;
}
````

### The CSS Box Model

![image](CSS%20Box%20Model.png)

Each element in a webpage can be visualised as being placed in a "box", whereby the dimensions of the box can be changed using width: and height: attributes (in percentage or pixels). 

<b>Margin Property</b><br>
The margin is the space outside of the border.

<b>Padding Property</b><br>
The padding property pushes the border out by a given value, without affecting the height and width of the box. Margin and Padding both also take either 2 or 4 values (as below)


<b>Border Property</b><br>
The box's border can also be changed using "border:" attribute, taking the values thickness (px), border style (solid/dashed), and color (in that sequence) e.g. border:30px solid blue. Changing the border width does not affect the box's height and width. Specific sides of the border can be edited e.g. border-top: 0px applied after preceding attribute removes the top border. "border-wdith:" can take four thicknesses from the top in a clockwise direction, e.g. border-width: 10px 15px 20px 25px. If only two values are provided, then the first value is applied to the top+bottom borders, while the second value is applied to the left and right borders.

In other words, the box contains the content, which can be modified using height and width, the padding is the space between the box and the border, the border itself can be modified, and the margin is the space between the border and other elements in the webpage. The specific attribute box model in inspection can be selected and edited.

<b>Grouping Boxes</b><br>
You can wrap a box around 2 or more elements by enclosing the elements within the html content division element "div". Chrome's [Pesticide](https://chromewebstore.google.com/detail/pesticide/bakpbgckdnepkmkeaiomhmfcnejndkbi) extension can show div-segregated element groups.
