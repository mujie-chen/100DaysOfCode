### Web Foundation Starts Here

Surfing the internet is basically interacting with an always-on web server from a client terminal. The browser sends a request message to the internet service provider (ISP), which liaises with the domain name server (DNS) to look up the [exact IP address](https://www.nslookup.io) of the requested website. Thereafter, you can make direct requests to that specific IP address for the content hosted there. Modern servers use dynamic IP addresses (changing based on usage) and a content distribution network (geographically distributed addresses to lower latency). Messages are sent across the world via optical fibres in [large underwater cables](https://www.submarinecablemap.com) at the speed of light.

#### HTML, CSS and JavaScript

To access a webpage, a browser looks up and receives data provided by a server. The data primarily consists of three code files: HTML, CSS and JavaScript.<br><br>
The <b>.HTML</b> establishes the qualitative structure of a website, such as images, buttons, or textboxes. Early websites, like the [first one to be created](https://info.cern.ch/hypertext/WWW/TheProject.html), contained HTML without any CSS or JavaScript. Conversely, however, it is not possible to have only the other two without a HTML file<br>
The <b>.CSS</b> stylises the aesthetics of the website, specifying the colour of the button, or the font of the text.<br>
The <b>.JavaScript</b> contains the functionality or behaviour of the website.

Then, the browser renders each of these files as a website

#### HyperText Markup Language

HyperText are pieces of text that can link to another documents in the website. [HTML tags](https://www.w3schools.com/TAGS/default.asp) ([Link 2](https://developer.mozilla.org/en-US/docs/Web/HTML/Guides/Cheatsheet)) are used to indicate the desired formatting to apply to the text. The tag + text is referred to as an element. Examples include the header element, in which the levels are inspired by book bindings, with top level headings and subsections (H1 through to H6), paragraph tags (for which [placeholder text](https://lipsum.com) are used to test formatting), unordered and ordered lists (with list items). There are also void elements, where content is forbidden within the tags, such as the horizontal separator "hr" and line break "br" tags (both enclosed in <>). Line breaks are used for natural new lines such as for poems, addresses etc.

#### HTML Boilerplate

HTML has several elements that are included with each document:
1. "!DOCTYPE html" declaration, which indicates to browsers the html version (version 5)
2. "html" start and end tags, the root of the document enclosing the entire document
3. html lang="en", the language of the content for screen readers
4. head, which includes important information about the website for rendering, but does not contain anything structurally. head contains the meta charset="UTF-8", the character set for encoding of the webpage, and title, the title displayed in the tab bar. meta name"viewport" content="width=device-width, initial-scale=1.0", which denotes the size of the window when the website is first opened.
5. body, in which the proper content of the website is nested (text, titeles, images, links, etc.)


#### HTML Attributes and the Anchor and Image Elements
HTML tags can take attributes. For example, in the anchor element, which is used to create hyperlinks, href="link.com" is added inside the <> after a. Each attribute is separated by a whitespace, but they all go inside the openign tag. Anchor then styles the text blue with underline. Certain attributes are elmenet-specific, while others are global and can be used by many elements (such as draggable=True). Similarly the img or image element uses the "src" attribute, taking a url to the image. img is a self-closing void element since there is no enclosed text. img also takes a alt attribute, which describes the image to blind people using screen readers.
